Hello everyone!

Today I will be analyzing a dataset in regards to the attrition rate of IBM employees (using a fictional dataset provided by IBM). The goal of this analysis is to come to a conclusion on as to what factors can be causing attrition and to provide solutions to the problem. I will be using basic matplotlib and seaborn to help visualize the data in order to quickly highlight any potential problems presented by the data set.

The first step we take is to import the csv file into our project, this will allow us to read the data and then extract the data we want to dissect later.

```python
import pandas as pd
import math
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")    # Ignore warnings

ibm = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
ibm.dropna(inplace=True)
ibm.head(10)
```

Having done that, our next step is to first identify how the severe the rate of attrition is for this data set by asking:

Question: How does IBM's rate of attrition compare to the national average rate of attrition and is it significant?

It is important to ask this question because it allows us and the company to gauge how severe the issue is, and if it even needs our immediate attention. So by using a bar chart we will display the attrition rate. We will also be using two distribution plots on the 'Age' to determine if at which age the Attrition rate is the highest.

```python
Attrition=ibm.Attrition.value_counts()
print(Attrition)
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.bar(x=Attrition.index, height=Attrition.values)
plt.title('Distribution of Employee Turnover')
plt.xlabel('Employee Turnover')
plt.ylabel('Count')

plt.subplot(1,2,2)
sb.distplot(ibm['Age'])
sb.distplot(ibm.Age[ibm.Attrition=='Yes'])
plt.legend(['Normal','Attrition: Yes'])
plt.show()
```

It can be seen by the bar chart on the left that this sample of IBM employees have an attrition rate of approximately 16.57% (237/1430 = .1657), this is roughly 6.57% higher then the "average" attrition rate of 10% provided by the national average (as of July 2018). Thus we can conclude that while the issue is not too extreme, it is definently worth our time to dig deeper into the problem.

The two distribution plots from the age tell us that around the ages of 25-35 is when the rate of attrition is highest, this is enlightening as usually the highest rate of attrition is found those in younger employees, typically around the age of 20-25. This can highlight that there are issues on IBM's ability to retain middle aged workers and can be worth a further investigation. 

Dealing with a large data set, it is important to look into smaller areas of our data first in order to better make sense and isolate potential outliers in our data. Therefore, ets take a look at the attrition rates based 'Departments' to see if there are any glaring issues. Here we asking the question:

Question: Is the rate of attrition normal throughout the diffferent departments of IBM employees? 

```python
plt.figure(figsize=(15,10))
plt.subplot(2, 1, 1)
sb.countplot(data=ibm, x="Department", hue="Attrition")
plt.title('Attrition based on Department')
plt.subplot(2, 1, 2)
sb.boxplot(x=ibm.DailyRate, y=ibm.Department)
plt.title('Daily Rate based on Department')
plt.show()
```

As can be seen on the graph above there is a rather large disparity in Attrition rate for Sales in comparison to the R&D Department and HR Department. While the R&D Department has much higher employee count it is important to note that it's attrition level is about 1/8 or 12.5% for it's department (which is acceptable as a roughly ~10% attrition is considered normal), whereas for Sales it is about 1/4 which is 25%. Sales has roughly half the amount of employees compared to R&D however it's Attrition rate is almost double that of R&D! This is a major concern and probably illustrates a deeper issue in the department itself rather then the company as a whole. Based on this knowledge, it would be wise to do further investigation on the Sales Department Employees and compare them to the other Departments to see if there are any disparities between them. 

Seeing how differences in departments have such a high effect on attrition rates, exploring to see if the disparity of pay for the employees in the Sales Department compared to those in other departments will allow us more insight as to the factors which cause such attrition. Using a box plot, I found that the average Daily Rate of a Sales employee is smaller then those of HR and R&D, however it does not have the lowest possible Daily Rate of the Departments. Thus this box plot indicates that the Daily Rate, while not the main factor, is likely contributing towards the high attrition rate of the Sales Department.

To further investigate payments in co-relation to attriton rate, we can take a look at the Stock Option Level to see if different options for company stocks affect the rate at which people stay at IBM. To this we ask the question:

Does having more stock options affect the rate of Attrition in IBM? How significant are the difference between 0 and 1 stock option?

```python
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
Stock=ibm.StockOptionLevel.value_counts()
plt.bar(x=Stock.index, height=Stock.values)
plt.title('Stock Option #')
plt.xlabel('Stock Option Level')
plt.ylabel('# of Employees')
plt.subplot(1, 2, 2)
sb.distplot(ibm.StockOptionLevel[ibm.Attrition=='No'])
sb.distplot(ibm.StockOptionLevel[ibm.Attrition=='Yes'])
plt.legend(['No','Yes'])
plt.title('Distribution of Attrition based on Stock Option Level')
plt.show()
```

The bar of the left is the distribution of stock options for all the ibm employees in this sample. As can be seen there is amount of employees that have no stock options compared to those with 1 stock option is roughly the same, with stock option levels of 2 and 3 to be much smaller. With that in mind we can now compare it to the graph on the right which shows the spread of attrition for the employees within those stock options. As can be seen, the attrition rate of people with no stock options is roughly 50% versus the very low attrition rate of level 1 and 2 stock option employees. There is roughly a 25% increase in employee retention when given atleast one stock option from no stock options. This is not surprising as giving stock options is a great incentive to stay with a company as well as incentivizes employees to care more about their work. Seeing such a drastic increase in retention rate based on providing a stock option however, could highlight a possible rememdy to the high attrition rate of this sample by simply being more lenient as the the criteria in which IBM provides their stock options to employees.

In conclusion, the rate of attrition in IBM while not severe is significant. The most glaring problem can be seen in it's sales department and while further analysis is required to determined what exactly is the underlying cause for the attrition rate, it can be safely determined that average daily rate and stock options for their employees affect the attrition rate.
