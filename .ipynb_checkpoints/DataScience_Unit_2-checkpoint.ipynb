```python
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('un.csv')
#df.dropna(inplace=True)
df.head()
```

```python
plt.scatter(x=df['lifeMale'],y=df['GDPperCapita'])
plt.scatter(x=df['lifeFemale'],y=df['GDPperCapita'])
plt.show()
#This scatter chart will allow us to see the effects of the GDP per Capita on life expectancy. As can be seen
#while you do not need a higher GDP to acquire higher life expectancy most of the time, in general a higher
#GDP will always result in a higher average life expectancy in both female and male.
```

```python
plt.hist(df['lifeMale'], color='blue',bins=np.arange(30, 80),alpha=.5)
plt.hist(df['lifeFemale'],color='red',bins=np.arange(30, 80),alpha=.5)
plt.show()
#This histogram chart allows us to visualize the general life expectancy of both male and females. What this
#allows us to understand is the general life expectancy averages of both male and female, and that females
#based on this data set have a higher life expectancy then males
```

```python
df.dropna(inplace=True)
plt.boxplot(df['GDPperCapita'])
plt.show()
#This box plot allows us to see where the majority of our data set lies, as can be seen the majority of our dataset
#resides within a GDPperCapita range of 2000-7000 with the mean at around 2500. This graph allows us to understand where
#we have the most data and can possibly show us where our assumptions will be the strongest/weakest.
```

```python
plt.scatter(df['infantMortality'], df['GDPperCapita'])
plt.show()
#This is a scatter plot on the infant mortality rate. As can be seen a higher GDP per Capita will very likely decrease infant
#mortality rate, in comparison to a lower GDP. While there are a few deviations, we can use this graph to highlight said 
#deviations and possibly use that information to highlight possible areas where improvment in infant care would be necessary. 
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.scatter(x=df['lifeMale'],y=df['GDPperCapita'])
plt.scatter(x=df['lifeFemale'],y=df['GDPperCapita'])
plt.ylabel('GDP per Capita')
plt.xlabel('Life of Male/Female')
plt.title('GDP on Life Expectancy')


plt.subplot(1, 2, 2)
plt.hist(df['lifeMale'], color='blue',bins=np.arange(30, 80),alpha=.5)
plt.hist(df['lifeFemale'],color='red',bins=np.arange(30, 80),alpha=.5)
plt.xlabel('Life Expectancy')
plt.ylabel('# of Occurances')
plt.title('Life Expectancy Based on Gender')
plt.show()

plt.figure(figsize=(10, 5))

plt.subplot(2,2,1)
df.dropna(inplace=True)
plt.boxplot(df['GDPperCapita'])
plt.ylabel('GDPperCapita')
plt.title('Average GDP per Capita')

plt.subplot(2,2,2)
plt.scatter(df['infantMortality'], df['GDPperCapita'])
plt.xlabel('Infant Mortality Rate')
plt.ylabel('GDP per Capita')
plt.title('GDP Effects on Infant Mortality')
plt.show()
#Subplot excercise using the plots provided from above
```
