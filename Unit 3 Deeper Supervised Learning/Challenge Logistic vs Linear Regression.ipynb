# Resample

```python
rawdata['RobberyExists'] = rawdata['Robbery'].apply(lambda x: 0 if x == 0.0 else 1)
print (rawdata['RobberyExists'].value_counts())
```

```python
target_count = rawdata.RobberyExists.value_counts()
print('Class 0:', target_count[0])
print('Class 1:', target_count[1])
print('Proportion:', round(target_count[0] / target_count[1], 2), ': 1')

target_count.plot(kind='bar', title='Robbery Spread');
```

```python
from sklearn.utils import resample
# Separate majority and minority classes
data_majority = rawdata[rawdata.RobberyExists==1]
data_minority = rawdata[rawdata.RobberyExists==0]
 
# Upsample minority class
data_minority_upsampled = resample(data_minority, 
                                 replace=True,     # sample with replacement
                                 n_samples=212)    # to match majority class
 
# Combine majority class with upsampled minority class
data_upsampled = pd.concat([data_majority, data_minority_upsampled])
 
# Display new class counts
data_upsampled.RobberyExists.value_counts()
```

```python
data_upsampled.head(10)
```

```python
target = data_upsampled['RobberyExists'].copy()
#Create our dependent variables, dropping RobberyExists (becuase its our target), Robbery, and Rape1(because it is all NaNs)
selected_data = data_upsampled.drop(['RobberyExists', 'Robbery', 'Rape1', 'City'], axis=1)
```

```python
#Because we are doing logistic regressions, it is imperative that we standardize the data 
#to allow for better processing (more info at https://medium.com/@rrfd/standardize-or-normalize-examples-in-python-e3f174b65dfc)
from sklearn import preprocessing
# Create the Scaler object
scaler = preprocessing.StandardScaler()
# Fit your data on the scaler object
selected_data = scaler.fit_transform(selected_data)
```

```python
from sklearn.linear_model import LogisticRegression
# Set up the regression model to predict defaults using all other
# variables as features.
regr = linear_model.LinearRegression()
regrfit = regr.fit(X_train, y_train)

# Calculate accuracy. First, get probability that each row will be admitted.
pred_statsmod = regrfit.predict(X_train)

# Code admission as 1 if probability is greater than .5.
pred_y_statsmod = np.where(pred_statsmod < .5, 0, 1)

# Accuracy table.
table = pd.crosstab(y_train, pred_y_statsmod)

print('\n Accuracy by admission status')
print(table)
print('\n Percentage accuracy')
print((table.iloc[0,0] + table.iloc[1,1]) / (table.sum().sum()))

print('\n Original Regression Score')
print(regrfit.score(X_test, y_test))
```

My understanding so far:
Linear vs Logistic Regression:
    Linear Regression is used when our dependent variable is a continuous variable and Logistic Regression is used when our dependent variable is a categorical variable.

Lasso vs Ridge Regression: 
    Ridge Regression: Ridge Regression is very similar to linear regression, however it's difference lies in it's regularization. It regulizes this by setting a lambda (or alpha) and if the coefficient exceeds this lambda, then it's weight on the overall model is reduced. This is important because effectively what it is doing is that it helps trim down on overfitting by preventing outliers from influencing the data.
    Lasso Regression: Lasso Regression is like Ridge Regression in purpose (to prevent overfitting) however it regulizes the data differently. Instead of punishing outliers, Lasso Regression will effectively set all outliers beyond a certain lambda (alpha) to 0, and reduces the data size along with it. This in effect simplifies the data which helps prevent overfitting.

Standardize data: When you’re running models (logistic regression, SVMs, perceptrons, neural networks etc.) the estimated weights will update similarly rather than at different rates during the build process. Standardizing the data helps smoothen out the process.
